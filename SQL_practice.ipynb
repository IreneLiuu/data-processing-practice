```python
import sqlite3
import pandas as pd
connection = sqlite3.connect("chinook.sqlite")
```

```python
cursor = connection.cursor()
```

```python
rows = cursor.execute("select title from albums limit 10").fetchall()
display(rows)

#splay(cursor.execute("select title from albums limit 10").fetchall()) #same effect as above code
```

```python
rows = cursor.execute("SELECT trackid, name, composer, unitprice FROM tracks;").fetchall()
# display(rows)

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
# df.head()
display(df)
```

```python
rows = cursor.execute("SELECT name, milliseconds, albumid FROM tracks ORDER BY albumid ASC, milliseconds DESC").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## SQL Order By
rows = cursor.execute("""
SELECT 
    TrackId, 
    Name, 
    Composer 
FROM 
   tracks
ORDER BY 
   Composer;
""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## SQL Select Distinct
rows = cursor.execute("""
SELECT  DISTINCT
	city,
	country
FROM
	customers
ORDER BY
	country;""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## SQL WHERE
rows = cursor.execute("""
SELECT
	name,
	milliseconds,
	bytes,
	albumid
FROM
	tracks
WHERE
	albumid = 1
AND milliseconds > 250000;""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)

rows = cursor.execute("""
SELECT
	name,
	albumid,
	composer
FROM
	tracks
WHERE
	composer LIKE '%Smith%'
ORDER BY
	albumid;""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
rows = cursor.execute("""
SELECT
	TrackId, 
	Name, 
	AlbumId
FROM
	Tracks
WHERE
	AlbumId IN (
		SELECT
			AlbumId
		FROM
			Albums
		WHERE
			ArtistId = 12);
                      """).fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
rows = cursor.execute("""
SELECT 
    Title,
    Name
FROM 
    albums
INNER JOIN artists 
    ON artists.ArtistId = albums.ArtistId;
                      """).fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```
