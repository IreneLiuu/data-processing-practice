```python
## SQL Order By
rows = cursor.execute("""
SELECT 
    TrackId, 
    Name, 
    Composer 
FROM 
   tracks
ORDER BY 
   Composer;
""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## SQL Select Distinct
rows = cursor.execute("""
SELECT  DISTINCT
	city,
	country
FROM
	customers
ORDER BY
	country;""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## SQL WHERE
rows = cursor.execute("""
SELECT
	name,
	milliseconds,
	bytes,
	albumid
FROM
	tracks
WHERE
	albumid = 1
AND milliseconds > 250000;""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)

rows = cursor.execute("""
SELECT
	name,
	albumid,
	composer
FROM
	tracks
WHERE
	composer LIKE '%Smith%'
ORDER BY
	albumid;""").fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
rows = cursor.execute("""
SELECT
	TrackId, 
	Name, 
	AlbumId
FROM
	Tracks
WHERE
	AlbumId IN (
		SELECT
			AlbumId
		FROM
			Albums
		WHERE
			ArtistId = 12);
                      """).fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
rows = cursor.execute("""
SELECT 
    Title,
    Name
FROM 
    albums
INNER JOIN artists 
    ON artists.ArtistId = albums.ArtistId;
                      """).fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
### SQL Group By
##  aggregate function such as MIN, MAX, SUM, COUNT, or AVG to provide more information about each group
rows = cursor.execute("""
SELECT
	albumid,
	COUNT(trackid)
FROM
	tracks
GROUP BY
	albumid
ORDER BY COUNT(trackid) DESC;
                      """).fetchall()

columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## GROUP BY with HAVING clause + INNER JOIN
rows = cursor.execute("""
SELECT
	tracks.albumid,
	title,
	COUNT(trackid)
FROM
	tracks
INNER JOIN albums ON albums.albumid = tracks.albumid
GROUP BY
	tracks.albumid
HAVING COUNT(trackid) > 15;
                      """).fetchall()
columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)

## SUM function
rows = cursor.execute("""
SELECT
	albumid,
	SUM(milliseconds) length,
	SUM(bytes) size
FROM
	tracks
GROUP BY
	albumid;
                      """).fetchall()
columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)
```

```python
## INSERT INTO function 
#FORMULA: 
# INSERT INTO table1 (column1,column2 ,..)
# VALUES 
#    (value1,value2 ,...),
#    (value1,value2 ,...),
#     ...
#    (value1,value2 ,...);

rows = cursor.execute("""
INSERT INTO artists (name)
VALUES
	("Buddy Ana");
                      """)

## verify the result using select+ order by + limit
rows = cursor.execute("""
SELECT
	ArtistId,
	Name
FROM
	artists
ORDER BY
	ArtistId DESC
LIMIT 3;
                      """).fetchall()
columns = [column[0] for column in cursor.description]
df = pd.DataFrame(rows, columns=columns)
display(df)

## commit function to save new row of data into orginal table >> connection.commit() or use autocommit = True
```

```python
# create a new table - artists_backup
rows = cursor.execute("""
CREATE TABLE artists_backup(
   ArtistId INTEGER PRIMARY KEY AUTOINCREMENT,
   Name NVARCHAR
);
                      """).fetchall()

## Inserting new rows with data provided by a SELECT statement(another table)
rows = cursor.execute("""
INSERT INTO artists_backup 
SELECT ArtistId, Name
FROM artists;
                      """)
```

```python
## verify
rows = cursor.execute("""
SELECT * FROM artists_backup;                     
                      """).fetchall()
display(rows)
# columns = [column[0] for column in cursor.description]
# df = pd.DataFrame(rows, columns=columns)
# display(df)
```

```python
## DELETE function
## delete artists whose names contain the word Santana
rows = cursor.execute("""
DELETE FROM artists_backup
WHERE name LIKE '%Santana%'
or    name =  "Buddy Rich" or "Candido" or "Charlie Byrd";                     
                      """)
rows = cursor.execute("""
select *
from artists_backup
limit 10;                     
                      """).fetchall()
display(rows)

## To remove all rows in the artists_backup table, you just need to omit the WHERE clause:
## >> DELETE FROM artists_backup;
```
